```python

from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options

options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches",["ignore-certificate-errors"])
options.add_argument('--disable-gpu')
options.add_argument('--headless')
# url = "https://www.facebook.com/ads/library/?active_status=all&ad_type=all&country=IN&view_all_page_id=110450027310691"
url = "https://www.facebook.com/ads/library/?active_status=all&ad_type=all&country=IN&view_all_page_id=189166971214680"
driver = webdriver.Chrome(executable_path = "C:\\Users\\Hemangi.Bavasiya\\Desktop\\Backups\\Machine Learning\\Chrome Driver\\chromedriver", options = options)



def scrapping (url):
    driver.get(url, )
    driver.refresh()
    driver.maximize_window() 
    name = driver.find_element_by_xpath('//div[@class="egkesoaz"]//child::div//child::a').text
    id1 = driver.find_element_by_xpath("(//div//child::div[@class='i0ppjblf e946d6ch'])[1]").text
    like = driver.find_element_by_xpath("(//div//child::div[@class='i0ppjblf e946d6ch'])[2]").text
    headingName = driver.find_element_by_xpath("//div[@class='_8n-w']//span[@role='heading']").text
    result = driver.find_element_by_xpath("//div[@class='_7gn2']//child::div").text
    launch = driver.find_element_by_xpath("//div[@class='_99s9']//child::div").text
    reJson = {}
    reJson['name'] = name
    reJson['page_ID'] = id1
    reJson['like'] = like
    reJson['header_name'] = headingName
    reJson['result'] = result
    reJson['launched'] = launch
    return reJson

data = scrapping(url)

print(data)
```
